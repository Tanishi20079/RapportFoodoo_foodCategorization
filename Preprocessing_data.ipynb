```python
path=r"/content/dataset"
# outPath="/content/drive/MyDrive/CGAS_FoodCategorization/DATA_IMAGES"
outPath="/content/drive/MyDrive/CGAS_DATA"
#Fetch From The Directory Path
dirs = os.listdir( path )
print(dirs)
dict_cat={"0":"gnocchi","1":"macarons","2":"french_toast","3":"donuts","4":"samosa","5":"dumplings","6":"waffles","7":"chocolate_cake","8":"sushi","9":"grilled_cheese_sandwich","10":"mussels","11":"hamburger","12":"fried_rice","13":"chocolate_mousse","14":"pancakes","15":"pizza","16":"ice_cream","17":"garlic_bread","18":"strawberry_shortcake","19":"hot_dog","20":"french_fries","21":"cup_cakes","22":"spring_rolls","23":"red_velvet_cake","24":"nachos" }

def resize():
    c=0
    for i in dirs:
        ff=path+"/"+i
        try:
          os.makedirs(outPath+"/"+i)
        except:
          pass
        di=os.listdir(ff)
        for file in di:
            print(file)
            #Getting The Full Path  
            filePaths=ff+"/"+file
            print(filePaths)

            #Fetch The Images
            im = Image.open(filePaths)

            #Get the Extension And Name Heres
            file_Name,extension = os.path.basename(filePaths).split(".")
            print(file_Name)
    #         file_Name
            #Set the ANTIALIAS for the Inverse Resizing Of Image
            imResize = im.resize((192,192), Image.ANTIALIAS)
            #Quality Can Be ALtered Depending On The Requirements

    #         
            #Save To The OtPath

            print(l[c]+"_"+file_Name)
            imResize.save(outPath+"/"+l[c]+"/"+(l[c]+"_"+file_Name)+'.'+extension, quality=100)
        c+=1
        print(c)
        


resize()
print("Preprocessing Heres.................... ")
```
